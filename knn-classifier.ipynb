```python
# import libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

```python
# Read dataset
cancer_df = pd.read_csv("data.csv").dropna(axis=1)
cancer_df.describe()
```

```python
# Pre-process dataset
X = cancer_df.drop(["id", "diagnosis"], axis=1)
encoder = LabelEncoder()
# Target feature is diagnosis
y = encoder.fit_transform(cancer_df["diagnosis"])
# Split into train and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# test different number of neighbors from 2 to 9
neighbors_sizes = np.arange(2, 10)
# save accuracies
model_accuracies = []
for n in neighbors_sizes:
  # Initialize and fit model
  model = KNeighborsClassifier(n_neighbors=n)
  model.fit(X_train, y_train)
  # Calculate accuracy
  accuracy = model.score(X_test, y_test)
  model_accuracies.append(accuracy)

# Find the index of model with best accuracy
best_idx = np.argmax(model_accuracies)
# Show the number of neighbors and the accuracy of the best model
print(f"KNN Classifier with best accuracy ({neighbors_sizes[best_idx]} neighbors): {model_accuracies[best_idx]}")
```
